```python
from sklearn.cluster import MiniBatchKMeans
import matplotlib.pyplot as pt
import numpy as np
```

```python
from forest import config
from forest.data_preprocessing import preprocess, split_features_target
from forest.dataset_split import create_sets
```

```python
df = preprocess(
    config.raw_path,
    config.get_features,
    config.dummies,
    config.drop_list
)
```

```python
X, y = split_features_target(df, 'is_fraud')
```

```python
X_train, X_test, y_train, y_test = create_sets(X, y, 0.08)
```

```python
X_train
```

```python
X_train_downsampled = X_train.sample(1000000, random_state=43)
```

```python
k_values = [25000, 50000]
centroids_dict = {}

for k in k_values:
    mbkm = MiniBatchKMeans(
        n_clusters=k,
        batch_size=10000,
        n_init="auto",
        random_state=42
    )
    
    mbkm.fit(X_train_downsampled)
    print(f"haha {k}")
    centroids_dict[k] = mbkm.cluster_centers_
```
